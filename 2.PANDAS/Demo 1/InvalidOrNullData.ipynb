```python
import pandas  as pd

df = pd.read_csv("Stock.csv",na_values={
    'eps':['Not Available','na'],
    'revenue':[-1,'na'],
    'people':['','na']
})
```

```python
def convert_price_cell(cell):
    if cell=="na":
        return 50
    elif cell < 0:
        return 0
    return cell

def convert_people_cell(cell):
    if cell=="na":
        return 'Unknown'
    return cell

df = pd.read_excel("StockData.xlsx","Sheet1",converters={
    'people':convert_people_cell,
    'price':convert_price_cell
})
df
```
